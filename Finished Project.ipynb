# Regression algorithms

```python
# shuffle the data since we want to use validation_curve 
car5=car4.sample(frac=1,random_state=0)
car5.head(2)
```

```python
# creating features and target sets
# use car3 to be our data 
x_car,y_car=car5.iloc[:,1:],car5['price']
display(x_car.head(2))
display(y_car.head(2))
```

```python
x_car.describe().round(2)
```

```python
car4.info()
```

```python
# encode catagorical features 
x_car_dummies=pd.get_dummies(x_car)
x_car_dummies.head(1)
```

```python
x_car_dummies.describe()
```

```python
# import algorithms
import pandas as pd 
import numpy as np
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.model_selection import cross_val_score, train_test_split,validation_curve
from sklearn.linear_model import LinearRegression,Ridge,Lasso,LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor,export_graphviz
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
```

```python
# split the data to train and test (using x_dummies instead of x_car!)
x_train,x_test,y_train,y_test=train_test_split(x_car_dummies,y_car,random_state=0)
```

## KNN regression 

```python
# use validation_cruve to find the best parameter value of n_neighbors
train_scores1,test_scores1 = validation_curve(KNeighborsRegressor(),x_train,y_train,param_name='n_neighbors',param_range=[2,3,4,5,9,15],cv=5)
```

```python
print('training:',train_scores1.mean(axis=1).round(4))
print('test:',test_scores1.mean(axis=1).round(4))
# n_neighbors 3 is the best parameter
```

```python
# define and fit the model using n_neighbors=4
knn_reg3=KNeighborsRegressor(n_neighbors=4)
knn_reg3.fit(x_train,y_train)
```

```python
# this is the result that should be reported in the project and presentation
# the test score doesn't look great but professor said it's okay
# we need to use all the methods we've learned to find out which one can give us best test scores!!!
print('knn_reg3 accuracy on training: {:.2%}'.format(knn_reg3.score(x_train,y_train)))
print('knn_reg3 accuracy on test: {:.2%}'.format(knn_reg3.score(x_test,y_test)),'\n')
```

## linear regression

```python
# define and fit the model 
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
# evaluate the model's performance
print('lr R2 on train:{:.2%}'.format(lr.score(x_train,y_train)))
print('lr R2 on test:{:.2%}'.format(lr.score(x_test,y_test)))
```

## ridge regression

```python
scaler1=MinMaxScaler()
x_car_tran=scaler1.fit_transform(x_car_dummies)
x_car_tran
```

```python
x_train1,x_test1,y_train1,y_test1=train_test_split(x_car_tran,y_car,random_state=0)
```

```python
# use validation_cruve to find the best parameter value of alpha
train_scores3,test_scores3 = validation_curve(Ridge(),x_train1,y_train1,param_name='alpha',param_range=[.001,.01,.05,.1,1,10],cv=5)
```

```python
print('training:',train_scores3.mean(axis=1).round(4))
print('test:',test_scores3.mean(axis=1).round(4))
```

```python
# define and fit the model
ridge= Ridge(alpha=1)
ridge.fit(x_train1,y_train1)
```

```python
# evaluate the performance
print ('ridge R2 on train:{:.2%}'.format(ridge.score(x_train1,y_train1)))
print ('ridge R2 on test:{:.2%}'.format(ridge.score(x_test1,y_test1)))
```

## lasso regression

```python
x_train1
```

```python
# use validation_cruve to find the best parameter value of alpha
# have too many categorical features, so it's okay to do have this error
train_scores4,test_scores4 = validation_curve(Lasso(),x_train1,y_train1,param_name='alpha',param_range=[.0001,.001,.01,1,5,10],cv=5)
```

```python
print('training:',train_scores4.mean(axis=1).round(4))
print('test:',test_scores4.mean(axis=1).round(4))
```

```python
# define and fit the model
lasso=Lasso(alpha=1)
lasso.fit(x_train1,y_train1)
```

```python
# evaluate the performance
print ('lasso R2 on train:{:.2%}'.format(lasso.score(x_train1,y_train1)))
print ('lasso R2 on test:{:.2%}'.format(lasso.score(x_test1,y_test1)))
print ('# of non-zero coefficients:',(sum(lasso.coef_!=0)))
```

## decision tree

```python
train_scores5,test_scores5=validation_curve(DecisionTreeRegressor(random_state=0),x_train,y_train,
                                        param_name='max_leaf_nodes',param_range=[16,17,18,19,20,50,100],cv=5)
```

```python
train_scores6,test_scores6=validation_curve(DecisionTreeRegressor(random_state=0),x_train,y_train,
                                        param_name='max_depth',param_range=[2,3,4,5],cv=5)
```

```python
print('training:',train_scores5.mean(axis=1).round(4))
print('test:',test_scores5.mean(axis=1).round(4))
```

```python
print('training:',train_scores6.mean(axis=1).round(4))
print('test:',test_scores6.mean(axis=1).round(4))
```

```python
dt_reg2=DecisionTreeRegressor(max_depth=5,random_state=0)
dt_reg2.fit(x_train,y_train)
```

```python
print('dt_reg2 R2 on train: {:.2%}'.format(dt_reg2.score(x_train,y_train)))
print('dt_reg2 R2 on test: {:.2%}'.format(dt_reg2.score(x_test,y_test)))
```
